```python
char_to_num = dict()
char_to_num['.'] = 0
for i in range(ord('z') - ord('a') + 1):
    char_to_num[chr(i + ord('a'))] = i + 1

num_to_char = dict()
for char, num in char_to_num.items():
    num_to_char[num] = char
```

```python
char_to_num, num_to_char
```

```python
import torch
```

```python
counts = torch.ones((27, 27), dtype=torch.int32)
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        counts[char_to_num[ch1], char_to_num[ch2]] += 1

counts
```

```python
P = torch.ones((27, 27), dtype=torch.float32)
for row in range(len(counts)):
    for col in range(len(counts[row])):
        P[row, col] = counts[row, col].float() / counts[row].float().sum()
```

```python
xs, ys = [], []

for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = char_to_num[ch1]
        ix2 = char_to_num[ch2]
        xs.append(ix1)
        ys.append(ix2)

xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
import torch.nn.functional as F
xenc = F.one_hot(xs, num_classes=27).float()
```

```python
g = torch.Generator().manual_seed(2147483647)
W = torch.randn((27, 27), generator=g, requires_grad=True)
logits = xenc @ W
counts2 = logits.exp()
probs = counts2 / counts2.sum(1, keepdim=True)
probs
```
